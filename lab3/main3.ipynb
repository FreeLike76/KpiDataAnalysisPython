# Task 4

## Generating data

```python
# shared column
name_id = np.random.randint(0, 1000, size=(150))
```

```python
df1 = pd.DataFrame(index=range(0, 100))
df1["name_id"] = name_id[0:100]
df1["weight"] = np.random.normal(3, 1, size=(100))
df1["height"] = np.random.normal(50, 2, size=(100))
df1
```

```python
df2 = pd.DataFrame(index=range(50, 150))
df2["name_id"] = name_id[50:150]
df2["weight_at_10"] = np.random.normal(32, 2, size=(100))
df2["height_at_10"] = np.random.normal(135, 10, size=(100))
df2
```

## Merge: Inner

```python
df_inner = pd.merge(df1, df2, how="inner", left_index=True, right_index=True, suffixes=(None, "_copy")).drop("name_id_copy", axis=1)
df_inner
```

## Merge: Left

```python
df_left = pd.merge(df1, df2, how="left", left_index=True, right_index=True, suffixes=(None, "_copy")).drop("name_id_copy", axis=1)
df_left
```

## Merge: Right

```python
df_right = pd.merge(df1, df2, how="right", left_index=True, right_index=True, suffixes=(None, "_copy")).drop("name_id_copy", axis=1)
df_right
```

## Merge: Outer

```python
df_outer = pd.merge(df1, df2, how="outer", left_index=True, right_index=True, suffixes=(None, "_copy")).drop("name_id_copy", axis=1)
df_outer
```
